```python
import darknet as dn
```

```python
dn.set_gpu(0)
net = dn.load_net("/home/ubuntu/darknet/cfg/yolov2.cfg", "/home/ubuntu/darknet/yolov2.weights", 0)
meta = dn.load_meta("/home/ubuntu/darknet/cfg/coco.data")
```

```python
def handlelabel(L):
    if len(L) == 0:
        return False
    for item in L:
        if (item[0] == 'car' or item[0] == 'truck') and item[1] > 0.8:
            return True
    return False
```

```python
files = os.listdir('/home/ubuntu/autoimages/autov2/img/t2/')
```

```python
files = files[10:]
```

```python
for folder in files:
    path = '/home/ubuntu/autoimages/autov2/img/t2/'+folder
    pics = os.listdir(path)
    for pic in pics:
        if pic[-3:] == 'jpg':
            direct = path+'/'+pic
            print(direct)
            yoloout = dn.detect(net,meta,direct)
            if not handlelabel(yoloout):
                os.remove(direct)
```
