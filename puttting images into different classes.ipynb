```python
for i in T[2:]:
    with open(i, "r") as read_file:
        data = json.load(read_file)
        if 'Doors' in data:
            if data['Doors']=='4':
                x='/home/dupad/autoimages/img/'+i[:-5]+'.jpg'
                y='/home/dupad/autoimages/4/'+i[:-5]+'.jpg'
                copyfile (x,y)
            elif data['Doors']=='2':
                x='/home/dupad/autoimages/img/'+i[:-5]+'.jpg'
                y='/home/dupad/autoimages/2/'+i[:-5]+'.jpg'
                copyfile (x,y)
            elif data['Doors']=='3':
                x='/home/dupad/autoimages/img/'+i[:-5]+'.jpg'
                y='/home/dupad/autoimages/3/'+i[:-5]+'.jpg'
                copyfile (x,y)
            elif data['Doors']=='5':
                x='/home/dupad/autoimages/img/'+i[:-5]+'.jpg'
                y='/home/dupad/autoimages/5/'+i[:-5]+'.jpg'
                copyfile (x,y)
            elif data['Doors']=='-':
                x='/home/dupad/autoimages/img/'+i[:-5]+'.jpg'
                y='/home/dupad/autoimages/-/'+i[:-5]+'.jpg'
                copyfile (x,y)
            else :
                x='/home/dupad/autoimages/img/'+i[:-5]+'.jpg'
                y='/home/dupad/autoimages/else/'+i[:-5]+'.jpg'
                copyfile (x,y)
```
